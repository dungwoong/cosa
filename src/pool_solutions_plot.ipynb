```python
import os
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
with open('single_run_output.json', 'r') as f:
    data = json.load(f)
```

```python
def extract_main_dict(item):
    # currently, the auto-generated dict is the last key that gets added
    run_data = item.pop(list(item.keys())[-1])
    item.update(run_data)

for item in data:
    extract_main_dict(item)
```

```python
df = pd.DataFrame(data)
```

```python
pd.set_option('display.max_colwidth', 500)
df.sort_values(by='solution_number')[['factor_config', 'spatial_config', 'mip_objective', 'energy']]
```

```python
df['config_str'] = df.apply(lambda row: str(row['factor_config'] + row['spatial_config'] + row['outer_perm_config']), axis=1)
df.drop_duplicates(subset=['config_str'], inplace=True)
```

```python
sns.scatterplot(df, x='mip_objective', y='energy')
plt.title('Energy Used vs MIP Objective For a Sample Workload/Architecture')
plt.ylabel('Energy(uJ)')
plt.xlabel('MIP Objective')
```
